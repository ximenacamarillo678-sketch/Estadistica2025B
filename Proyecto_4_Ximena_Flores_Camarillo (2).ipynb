INTRODUCCIÓN
El rendimiento académico es un tema central en la investigación educativa, ya que representa uno de los principales indicadores del aprendizaje y del desarrollo estudiantil. Diversos factores pueden influir en el desempeño, desde condiciones familiares y socioeconómicas hasta características personales, escolares y del entorno. Comprender cuáles de estos factores tienen un impacto significativo permite a las instituciones diseñar estrategias más efectivas para apoyar a los estudiantes.
Este  proyecto tiene como objetivo analizar una serie de variables relacionadas con el rendimiento académico mediante el uso de técnicas estadísticas, específicamente el análisis de varianza (ANOVA) y las pruebas post hoc de Tukey. Estas herramientas permiten determinar si existen diferencias estadísticamente significativas entre grupos y, por lo tanto, identificar factores que realmente influyen en los puntajes obtenidos por los estudiantes.
Dentro de las variables estudiadas se encuentran elementos como la participación parental, el acceso a recursos educativos, la motivación, el nivel socioeconómico, la calidad del profesorado, la distancia al centro educativo, la influencia de los compañeros, la presencia de discapacidades de aprendizaje, el tipo de escuela, entre otras. Todas ellas se analizaron de forma independiente para determinar su efecto sobre el desempeño académico.
A lo largo del proyecto se pueden observar los fundamentos teóricos, la metodología utilizada, el resumen de los resultados obtenidos en cada factor y una conclusión general sobre la magnitud e importancia de dichas influencias. El análisis integral de estas variables permite comprender que algunos factores tienen un papel más relevante que otros y que, en conjunto, ofrecen una visión amplia de los elementos que pueden favorecer o limitar el aprendizaje.

El rendimiento académico ha sido ampliamente estudiado dentro del campo de la educación, ya que constituye un indicador esencial del aprendizaje y del desarrollo escolar. De acuerdo con distintos enfoques pedagógicos y psicológicos, el desempeño de un estudiante puede estar influido por factores personales, familiares, escolares y contextuales. Estos elementos interactúan entre sí y determinan, en mayor o menor medida, la capacidad del alumno para obtener resultados satisfactorios.

Uno de los factores más estudiados es la participación parental, entendida como el grado de acompañamiento, apoyo y supervisión que las familias brindan en el proceso educativo. La literatura señala que los estudiantes cuyos padres se involucran activamente suelen mostrar mayor motivación, mejor asistencia escolar y desempeño académico superior. Asimismo, la investigación ha mostrado que el acceso a recursos, como materiales educativos y tecnología, especialmente internet, facilita la realización de tareas, la búsqueda de información y el desarrollo de competencias académicas.

La motivación estudiantil es otro componente clave. Según las teorías motivacionales, los estudiantes con mayor interés, disposición y compromiso hacia el aprendizaje tienden a esforzarse más y obtener mejores resultados. Del mismo modo, el nivel socioeconómico y el nivel educativo de los padres suelen asociarse con oportunidades de aprendizaje más amplias, mayor disponibilidad de recursos académicos y entornos más favorables para el desarrollo escolar.

En el ámbito escolar, factores como la calidad del profesorado, el tipo de institución y la distancia entre el hogar y la escuela pueden influir en el desempeño académico. Una enseñanza de calidad, caracterizada por estrategias pedagógicas efectivas, retroalimentación adecuada y dominio del contenido, contribuye al aprendizaje significativo. Por último, variables personales como la presencia de una discapacidad de aprendizaje pueden alterar la forma en que el estudiante procesa, retiene o expresa la información, por lo que es esperable que se observen diferencias en el desempeño si no se cuenta con apoyos adecuados. En conjunto, estos factores conforman un marco conceptual que permite analizar de manera integral los elementos que potencialmente explican las variaciones en el rendimiento académico y fundamentan el estudio estadístico realizado.





















METODOLOGÍA



Para analizar los factores asociados al rendimiento académico, se aplicó una metodología cuantitativa basada en técnicas estadísticas inferenciales. El objetivo fue determinar si variables sociodemográficas, escolares y personales generaban diferencias significativas en los puntajes obtenidos por los estudiantes. El análisis se desarrolló utilizando un conjunto de datos que incluía diversas variables categóricas, cada una representada por niveles o grupos específicos.

El procedimiento principal consistió en aplicar un Análisis de Varianza (ANOVA) para cada variable. Esta técnica permite comparar las medias de distintos grupos y establecer si las diferencias observadas son estadísticamente significativas. Para cada ANOVA se examinó el valor p obtenido, utilizando como criterio de referencia un nivel de significancia de 0.05. Cuando el valor p fue menor a este umbral, se concluyó que existían diferencias significativas entre los grupos evaluados.

Después, se realizaron pruebas post hoc de Tukey con el fin de identificar entre qué grupos específicos se presentaban dichas diferencias. Este análisis complementario permitió detallar si las diferencias se daban entre niveles bajos, medios o altos de cada variable, así como determinar qué grupos presentaban los puntajes más elevados o bajos dentro del rendimiento académico.

Al último, se calculó el tamaño del efecto mediante el coeficiente eta cuadrado (η²), con el propósito de medir la magnitud real del impacto de cada variable sobre el rendimiento académico. Este indicador permitió complementar la significancia estadística con una interpretación práctica, revelando si los factores analizaban explicaban una proporción importante o reducida de la variabilidad en los puntajes.

En conjunto, esta metodología permitió obtener una evaluación integral, identificando tanto la existencia de diferencias significativas como la fuerza de la relación entre cada factor y el rendimiento académico, proporcionando así una visión completa del impacto de cada variable incluida en el estudio.







RESULTADOS




El análisis estadístico reveló que varios factores influyen significativamente en el rendimiento académico, aunque la magnitud de estos efectos varía considerablemente entre las variables evaluadas. En primer lugar, se encontró que la participación parental, el acceso a recursos educativos, la motivación, el nivel de ingreso familiar, la calidad del profesorado, la influencia de los compañeros, la presencia de discapacidad de aprendizaje, el nivel educativo de los padres y la distancia entre el hogar y la escuela mostraron diferencias estadísticamente significativas entre los grupos analizados. En todos estos casos, los valores p fueron menores a 0.05, lo que confirmó la existencia de diferencias reales entre los niveles evaluados.

Al analizar las comparaciones mediante la prueba de Tukey, se observaron patrones consistentes. En variables como participación parental, acceso a recursos, motivación, ingreso familiar y educación de los padres, los niveles más altos se asociaron con mejores puntajes académicos. En el caso de la calidad del profesorado, se encontraron diferencias únicamente cuando el nivel era muy elevado, mientras que para la distancia, se observó una tendencia negativa: a mayor distancia del hogar a la escuela, menor desempeño académico. Por su parte, la influencia positiva de los compañeros se vinculó con puntajes superiores, mientras que los estudiantes con discapacidad de aprendizaje mostraron promedios más bajos que quienes no la presentan.

Sin embargo, no todos los factores evaluados mostraron impacto en el rendimiento. Las variables tipo de escuela (pública o privada) y género no presentaron diferencias significativas entre sus grupos, y el tamaño del efecto en ambos casos fue prácticamente nulo, lo que indica que estas características no explican variación alguna en los puntajes.

En cuanto a la magnitud de los efectos, la mayoría de las variables presentó valores de eta cuadrado (η²) pequeños o muy pequeños, indicando que, aunque existe una relación estadísticamente significativa en varios casos, la proporción real de la variabilidad explicada en los puntajes es reducida. Ninguna variable individual mostró un efecto grande, lo que sugiere que el rendimiento académico es un fenómeno multifactorial influido por múltiples elementos que, de manera individual, tienen impactos limitados.





```python
import pandas as pd
url = "https://raw.githubusercontent.com/Estadistica2025B/Estadistica2025B/refs/heads/main/dataset.csv"
df = pd.read_csv(url)
df = df.dropna()
df
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Nivel de significancia = 0.05
# H0: Todas las medias son iguales
# H1: Al menos una media es distinta

# Realizar el modelo ANOVA de dos vías
modelo_lineal = ols(f'puntaje_examen ~ horas_estudio + asistencia + horas_suenio + calificaciones_previas + sesiones_tutoria + actividad_fisica + C(participacion_parental) + C(acceso_a_recursos) + C(actividades_extracurriculares) + C(nivel_motivacion) + C(acceso_internet) + C(ingreso_familiar) + C(calidad_profesorado) + C(tipo_escuela) + C(influencia_companieros) + C(discapacidad_aprendizaje) + C(nivel_estudio_padres) + C(distancia) + C(genero)', data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal, typ=2)
tabla_anova
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="participacion_parental", xlabel="puntaje_examen")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

```python
# η² (eta cuadrado)	Tamaño del efecto
# ~0.01	              Pequeño
# ~0.06	              Mediano
# ~0.14 o más	        Grande

# Se interpreta como el coeficiente de determinación

# Estos valores pueden variar según la disciplina. En estudios de ciencias
# sociales, un η² de 0.06 puede ser muy significativo;
# en física o ingeniería, quizá se espera más.

eta_squared = tabla_anova['sum_sq']['C(participacion_parental)'] / tabla_anova['sum_sq'].sum()
print(f"η² (tamaño de efecto): {eta_squared:.3f}")
```

El análisis mostró que la participación parental tiene un efecto significativo en el rendimiento académico. El ANOVA indicó diferencias claras entre los niveles de participación (p < 0.001), y las comparaciones de Tukey confirmaron que los tres grupos (bajo, medio y elevado) difieren significativamente entre sí, siendo el grupo con mayor participación el que obtuvo los puntajes más altos. Aunque el tamaño del efecto fue pequeño a cercano a mediano (η² = 0.030), lo que sugiere que la participación parental no explica una gran parte de la variabilidad total, sí contribuye de manera consistente y relevante al desempeño, por lo que se concluye que una mayor participación parental se asocia con mejores resultados académicos.

CONCLUSIÓN

El análisis realizado permitió identificar que diversos factores tienen un efecto estadísticamente significativo sobre el rendimiento académico; sin embargo, la magnitud de estos efectos suele ser reducida. Variables como la participación parental, el acceso a recursos educativos, la motivación del estudiante, el nivel de ingreso familiar, la calidad del profesorado, la influencia de los compañeros, la presencia de discapacidades de aprendizaje, el nivel educativo de los padres y la distancia entre el hogar y la escuela mostraron diferencias claras entre sus niveles, lo que confirma que todos estos elementos influyen en los puntajes obtenidos. No obstante, en la mayoría de los casos, los tamaños del efecto fueron pequeños, lo que indica que, aunque estas variables sí aportan a la explicación del rendimiento, no determinan por sí solas el desempeño académico.

Por otro lado, se constató que factores como el género y el tipo de escuela no presentan influencia significativa sobre el rendimiento, ya que los análisis ANOVA y las pruebas de Tukey no mostraron diferencias entre los grupos correspondientes. Estos resultados sugieren que no todas las características sociodemográficas o escolares tienen el mismo peso en el desempeño del estudiante, y que algunas creencias comunes sobre estas variables pueden no estar respaldadas por los datos.

En general, los resultados indican que el rendimiento académico es un fenómeno complejo y multifactorial. Ningún factor aislado explica una gran parte de la variación en los puntajes, sino que es la combinación de múltiples elementos lo que determina el desempeño de un estudiante. Esto resalta la importancia de abordar el aprendizaje desde una perspectiva integral, considerando tanto el entorno familiar como el escolar y personal del alumno.

Por ultimo, este estudio demuestra la utilidad del ANOVA y de las pruebas post hoc para identificar diferencias entre grupos y evaluar el impacto de diversos factores sobre una variable educativa. Aunque varios factores influyen en el rendimiento académico, sus efectos individuales son limitados, lo que abre la puerta a futuras investigaciones que analicen combinaciones de variables o modelos predictivos más complejos que permitan comprender mejor la interacción de estos elementos en el proceso de aprendizaje.
